```python
movie_cols = ["movie_id","title","release data","video release date","IMDB URL","unknown","Action","Adventure","Animation","Children's","Comedy","Crime","Documentary","Drama","Fantasy","Film-Noir","Horror","Musical","Mystery","Romance","Sci-Fi","Thriller","war","Western"]
movies = pd.read_csv("/content/drive/My Drive/data (1)/u.item",sep="|", names=movie_cols,encoding="ISO-8859-1")
movies = movies.set_index("movie_id")
movies.head(3)
```

```python
# u.data 데이터 불러오기
```

```python
rating_cols =["user_id","item_id","rating","timestamp"]
ratings=pd.read_csv("/content/drive/My Drive/data (1)/u.data", sep="\t",names=rating_cols)

display(ratings.head(5))
```

```python
# 평점 데이터(ratings)를 "movie_id"별로 묶고 평균 평점을 계산
movie_rating_mean = ratings.groupby(["item_id"])["rating"].mean()
movie_rating_mean
```

```python
# movie_rationg_mean 내림차순 정렬 후 5개만 추출

movie_rating_sort = movie_rating_mean.sort_values(ascending=False)[:5]

movie_rating_sort
```

```python
recom_movies = movies.loc[movie_rating_sort.index]
recom_movies
```

```python
recommendations=recom_movies["title"]
recommendations
```

```python
#정확도 계산을 위해  RMSE를 이용  
import numpy as np
def RMSE(y_true, y_pred): # y_true: 사용자가 실제 평가한 평점  , y_pred: 영화의 평균 평점
  return np.sqrt(np.mean((np.array(y_true)-np.array(y_pred))**2))

rmse = []
for user in set(ratings.index): #ratings 의 index 는 user_id , set의 의미? 중복을 허용하지 않는 집합 
  y_true = ratings.loc[user]["rating"] #user가 부여한 실제 평점
  y_pred = movie_rating_mean[ratings.loc[user]["item_id"]]  #user가 평가한 영화의 id 리스트를 추출 그리고 그것에대한 평균 평점
  accuracy = RMSE(y_true, y_pred)
  rmse.append(accuracy)

print(rmse)
```
